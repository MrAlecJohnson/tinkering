```python
use = copy.copy(test)
p1 = [int(i) for i in use[use.index("Player 1:") + 1:use.index("")]]
p2 = [int(i) for i in use[use.index("Player 2:") + 1:]]
```

```python
def play_round(deck1, deck2):
    card1 = deck1.pop(0)
    card2 = deck2.pop(0)
    #print(f"card1 = {card1}, card2 = {card2}")
    if card1 > card2:
        deck1.append(card1)
        deck1.append(card2)
    elif card2 > card1:
        deck2.append(card2)
        deck2.append(card1)
    #print(f"deck1 now = {deck1}")
    #print(f"deck2 now = {deck2}")

while p1 and p2:
    play_round(p1, p2)
    
print(p1)
print(p2)
```

```python
def score(deck):
    total = 0
    deck.reverse()
    for i, card in enumerate(deck, 1):
        total += i * card
    return total

print(score(p1))
print(score(p2))
```

```python
#### Part 2
# End if the same deck are back
#if both players' deck size >= card value, play subgame
# subgame uses copies
# otherwise winner is higher card
fail = [
    "Player 1:",
    43,
    19,
    "",
    "Player 2:",
    2,
    29,
    14,
]

use = copy.copy(data)
p1 = [int(i) for i in use[use.index("Player 1:") + 1:use.index("")]]
p2 = [int(i) for i in use[use.index("Player 2:") + 1:]]

def play_game(deck1, deck2, seen=None):
    if not seen:
        seen = []
    if deck1 in seen:
        #print("BREAK! Repeated deck1")
        return (1, deck1)
    else:
        seen.append(copy.copy(deck1))

    card1 = deck1.pop(0)
    card2 = deck2.pop(0)
    #print(f"card1 = {card1}, card2 = {card2}")
    if len(deck1) >= card1 and len(deck2) >= card2:
        #print("Starting subgame")
        result, deck = play_game(copy.copy(deck1[:card1]), copy.copy(deck2[:card2]), None)
        #print("Finishing subgame")
    else:
        if card1 > card2:
            result = 1
        elif card2 > card1:
            result = 2
    
    if result == 1:
        deck1.append(card1)
        deck1.append(card2)
        if len(deck2) == 0:
            #print("Player 1 wins")
            return (1, deck1)
    else:
        deck2.append(card2)
        deck2.append(card1)
        if len(deck1) == 0:
            #print("Player 2 wins")
            return (2, deck2)
        
    return play_game(deck1, deck2, seen)


result = play_game(p1, p2)
print(result)
print(score(result[1]))
```
