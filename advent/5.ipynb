```python
def ceiling(a: int, b: int):
    return -(-a // b)


def get_seat(seat: str):
    rows = seat[0:7]
    cols = seat[7:]
    max_row = 127
    min_row = 0
    max_col = 7
    min_col = 0
    for r in rows:
        diff = ceiling((max_row - min_row), 2)
        if r == "F":
            max_row -= diff
        elif r == "B":
            min_row += diff
        else:
            raise ValueError("Invalid row character")
    for c in cols:
        diff = ceiling((max_col - min_col), 2)
        if c == "L":
            max_col -= diff
        elif c == "R":
            min_col += diff
        else:
            raise ValueError("Invalid column character")
    return (min_row, min_col)


def get_id(seat: tuple):
    return (seat[0] * 8) + seat[1]


def convert(seat: str):
    number = get_seat(seat)
    return get_id(number)
```

```python
test = "BFFFBBFRRR" 
number = get_seat(test)
seat_id = get_id(number)

# row 70, column 7, seat ID 567
print(number)
print(seat_id)
```

```python
seen = [convert(d) for d in data]
top = max(seen)
print(top)
```

```python
find = sorted(seen)
bottom = min(find)

# clearly a sorting algo way of doing this, but...
for i in range(bottom + 1, top):
    if i not in find and (i - 1) in find and (i + 1) in find:
        print(i)
```
