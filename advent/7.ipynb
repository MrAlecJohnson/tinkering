```python
data = utils.get_local_input("inputs_7.txt", True)
data[0]
```

```python
colours = {}

for d in data:
    split_text = d.split(" ", 4)
    if split_text[4] != "no other bags.":
        outer = " ".join(split_text[:2])
        inner = set(colour.strip() for colour in re.sub(r"\d+ |bags?", "", split_text[4].rstrip(". ")).split(","))
        colours[outer] = inner
```

```python
answers = set()
def contained_by(target_colour, words):
    additions = set(c for c in colours if target_colour in colours[c])
    answers.update(additions)
    for a in additions:
        contained_by(a, words)

contained_by("shiny gold", colours)
```

```python
# Part 2
numbers = {}

for d in data:
    split_text = d.split(" ", 4)
    outer = " ".join(split_text[:2])
    numbers[outer] = numbers.get(outer, {})
    if split_text[4] != "no other bags.":
        cleaned = re.sub(" bags?", "", split_text[4]).rstrip(".").split(", ")
        for entry in cleaned:
            divided = entry.split(" ", 1)
            numbers[outer][divided[1]] = int(divided[0])

numbers
    
```

```python
def contains(target_colour, numbers=numbers):
    total = 0
    for colour in numbers[target_colour]:
        total += numbers[target_colour][colour] * (1 + contains(colour, numbers))
    return total
```

```python
contains("shiny gold", numbers)
```
